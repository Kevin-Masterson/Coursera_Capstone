```python
#get the data 
#Run the code from previous module
url = "https://en.wikipedia.org/wiki/List_of_Dublin_postal_districts"
dublin_hoods = pd.read_html(url, match="Dublin's postal districts")
print("list imported")
```

```python
#lets clean up the northside df
dubnorth.rename(columns={"Dublin's postal districts": "Area Code"}, inplace=True)
dubnorth.drop([0,14], inplace=True)
dubnorth.dropna(inplace=True)
dubnorth.reset_index(drop=True, inplace=True)
print(dubnorth.shape)
dubnorth.head(100)
```

```python
#I can see that some area codes have neighbourhoods in here as well so i will stick to the area code as it is more regular
dubnorth['Area Code'] = dubnorth['Area Code'].apply(lambda x: x.split(',')[0])
dubnorth.head(100)
```

```python
dubnorth['Latitude'] = None
dubnorth['Longitude'] = None

for i, area in enumerate(dubnorth['Area Code']):
    lat_lng_coords = None
    
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Dublin, Ireland'.format(area))
        lat_lng_coords = g.latlng
    
    if lat_lng_coords:
        latitude = lat_lng_coords[0]
        longitude = lat_lng_coords[1]
    
    dubnorth.loc[i, 'Latitude'] = latitude
    dubnorth.loc[i, 'Longitude'] = longitude

dubnorth.head(10)
```
